```python
import numpy as np

def F(x, w, b):
    result = np.dot(x,w)+ b
    return result
    
def J(x, y, w, b):
    m = len(x)
    yhat = F(x, w, b)
    cost = np.sum((y - yhat) ** 2) / (2 * m)
    return cost
    
def dJ_w(x, y, w, b):
    m = len(x)
    yhat = F(x, w, b)
    dJ_dw = -np.dot((y - yhat), x) / m
    return dJ_dw
    
def dJ_b(x, y, w, b):
    m = len(x)
    yhat = F(x, w, b)
    dJ_db = -np.sum(y - yhat) / m
    return dJ_db

def fmin(alpha, tol, x, w, b, y, mmax):
    err1 = 1
    itera = 0
    while(err1 > tol and itera < mmax):
        err1 = J(x, y, w, b)
        itera += 1
        print(f'n={itera};\t w={w} \t b={b} \t err = {err1}')
        w -= alpha * dJ_w(x, y, w, b)
        b -= alpha * dJ_b(x, y, w, b)
    return w, b

tol = 1e-4
alpha = 0.01

b = 0
x = np.array([0, 0.1, 1, 1.8, 2.5, 3.1, 3.3])
y = np.array([5, 4.7, 4, 3.7, 3.9, 4, 3.7])

w = np.random.randn(x.shape[0],1)
mmax = 800
w , b = fmin(alpha, tol, x, w, b, y, mmax)
```
